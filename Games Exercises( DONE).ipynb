#### Import pandas and read in the Games csv file and set it to a DataFrame called gms

```python
import pandas as pd
gms = pd.read_csv('games.csv')
```

#### Check the head of the DataFrame.

```python
gms.head()
```

####  Use .info() method to find out total number of entries in dataset

```python
gms.info()
```

#### How many rows and columns are present in the dataset?

```python
noOfRows=len(gms.axes[0])
noOfCols=len(gms.axes[1])
print("\033[1mNumber of ROWS in the dataset = ",noOfRows,"\nNumber of COLUMNS in the dataset = ",noOfCols)
# using \033[1m to make the output line bold (boldness not visible while viewing in git hub)
```

#### What is the highest number of comments received for a game?

```python
print("the highest number of comments received for a game = ",gms["total_comments"].max())
```

#### What is the name of the game with id 1500?

```python
nameWithId1500 = gms[(gms['id'] == 1500) ] 
nameWithId1500 = list(nameWithId1500["name"])  
#print("the name of the game with id 1500 : ",nameWithId1500)
#this list contains only one name so taking index zero make the otput proper
print("\033[1mthe name of the game with id 1500 : ",nameWithId1500[0])
```

#### Which games have received the least number of comments?

```python
leastComments = gms[(gms['total_comments'] == gms["total_comments"].min())] 
print("\033[1mName of the games which have received the least number of comments :")
print(list(leastComments["name"]))
```

#### Which games have received the highest number of comments?


```python
highestComments = gms[(gms['total_comments'] == gms["total_comments"].max()) ] 
print("\033[1mName of the games which have received the least number of comments :")
#print(list(highestComments["name"]))
#as it contains duplicate values using set to get the proper output
print(set(highestComments["name"]))

```

#### How many unique games are there in the dataset?

```python
print("\033[1mNumber of unique games = ", gms["name"].nunique())
```

#### How many boardgames and boardgameexpansions are there in the dataset?

```python
noOfBoardGames = gms[(gms.type == 'boardgame')]
print("\033[1mNumber of board games = ",len(noOfBoardGames))
noOfBoardGameExpansions = gms[(gms.type == 'boardgameexpansion')]
print("\033[1mNumber of board games = ",len(noOfBoardGameExpansions))
```

#### What is the mean playin time for all games put together ?

```python
print(" \033[1mthe mean playin time for all games put together is ",gms["playingtime"].mean())
```

#### Is there a correlation between playing time and total comments for the games? hint:- Use the .corr() function

```python
correlation = gms["playingtime"].corr(gms["total_comments"])
if(correlation):
    print("\033[1myes, there is a correlation between playing time and total comments for the games")
else:
    print("\033[1mNo, there is no correlation between playing time and total comments for the games")
print(correlation)
```
